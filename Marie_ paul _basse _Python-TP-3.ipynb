````python
##Exercice 31
""""
write a Python function that takes a list as input and returns a new list with all duplicates removed.
"""
def remove_duplicates_ordered(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list


original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = remove_duplicates_ordered(original_list)
print("Original List:", original_list)
print("Unique List (Ordered):", unique_list)                                                                                                                                                                                                                                            
````

````python
## Exercise 36:
""""
Write a Python function that takes a list of strings as input and returns the longest string.
"""
def find_longest_string(strings):
    if not strings:  
        return None
    return max(strings, key=len)

string_list = ["apple", "banana", "cherry", "blueberry"]
longest = find_longest_string(string_list)
print("The longest string is:", longest)
````

````python
### Exercise 40:
""""
Write a Python function that flattens a list of lists into a single list.
"""
def flatten_list(nested_list):
    flat = []
    for sublist in nested_list:
        for item in sublist:
            flat.append(item)
    return flat


liste = [[1, 2], [3, 4], [5]]
print(flatten_list(liste))  
````

````python
## Exercise 48:
""""
Write a Python function that uses a generator to yield the Fibonacci sequence up to `n` terms.
"""
def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci_generator(10):
    print(num)
````
